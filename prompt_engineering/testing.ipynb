```python
# Example: reuse your existing OpenAI setup
from openai import OpenAI

# Point to the local server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="not-needed")


prompt = """
You are John Lin, you are 25 years old. You live in Milan.
You are currently aware that a virus is spreading across the country. Scientists are warning of a potential epidemic. 
Currently, 0% of the population in Milan has been infected.
John Lin is not afraid of the virus spreading in the city. John Lin doesn’t care about spreading the virus to his contacts, and disregards the well-being of others.
our friends are: Luca, Tommaso.

Given your belief about the virus, you need to make the following decisions:
- Go to work OR stay at home
- If you go to work, do you take the public transport (cheap but could expose you to infected people) or car (expensive, but safer)?
- If you go to work, do you wear a mask or not?
- After work, you can decide if you want to participate in a social activity with your friends.
"""


completion = client.chat.completions.create(
  model="local-model", # this field is currently unused
  messages=[
    {"role": "system", "content": "Impersonate John Lin"},
    {"role": "user", "content": prompt}
  ],
  temperature=0.7,
)

print(completion.choices[0].message)
```
