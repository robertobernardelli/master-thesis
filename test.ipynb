```python
api_key = open('openai-api-key.txt').read().strip()
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)


agent_name = 'John Lin'
agent_age = 25

symptoms_status1 = f'{agent_name} has a fever and a cough.'
symptoms_status0 = f'{agent_name} is feeling good, no symptoms.'

symptoms_status_list = [symptoms_status0, symptoms_status1]

perc_pop_infected_list = ['0.0', '0.1', '0.2', '0.5', '1.0', '1.5', '2.0', '5.0', '10.0']

# store results in dict
results = {}

traits = ['Distrust, Indecisiveness, Unaggressiveness, Independence, Imperceptiveness',
    'Introspective, Curious, Original, Organized',
    'Outgoing, Unstable, Efficient, Wide Interests']

for traits in traits:
    for symptoms_status in symptoms_status_list:
        for perc_pop_infected in perc_pop_infected_list:
            prompt = f"""
            You are {agent_name}. You are {agent_age} years old. 
            Your traits are given below:
            
            Your basic bio is below:
            {agent_name} lives in the town of Dewberry Hollow. {agent_name} likes the town and has friends who also live there. {agent_name} has a job and goes to the office for work everyday.
            I will provide {agent_name}'s relevant memories here:
            {symptoms_status}
            {agent_name} knows about the Catasat virus spreading across the country. It is an infectious disease that spreads from human to human contact via an airborne virus. The deadliness of the virus is unknown. Scientists are warning about a potential epidemic.
            {agent_name} checks the newspaper and finds {perc_pop_infected}% of Dewberry Hollow's population caught new infections of the Catasat virus yesterday.
            Based on the provided memories, should {agent_name} stay at home for the entire day? Please provide your reasoning.
            If the answer is "Yes," please state your reasoning as "Reasoning: [explanation]." If the answer is "No," please state your reasoning as "Reasoning: [explanation]." The format should be as follow:
            Reasoning:
            Response:

            Example response format:
            Reasoning: {agent_name} is tired. 
            Response: Yes

            It is important to provide Response in a single word.
            """

            response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": f"Impersonate {agent_name}"},
                {"role": "user", "content": prompt},
            ]
            )

            response.choices[0].message.content
            results[(symptoms_status, perc_pop_infected, traits)] = response.choices[0].message.content
            print(f'{symptoms_status}, {perc_pop_infected}, {traits}')
            print(response.choices[0].message.content)
```

```python
# save results as txt
with open('results.txt', 'w') as f:
    for key, value in results.items():
        f.write('%s:%s\n' % (key, value))
        f.write('####################\n')
```
